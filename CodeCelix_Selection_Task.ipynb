```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

st.title("AI Customer Support System (Fully AI-Based) By Sardar Ammer Qadeer ")


# Load models (only once)
@st.cache_resource
def load_models():
    classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
    sentiment = pipeline("sentiment-analysis")
    return classifier, sentiment

classifier, sentiment_model = load_models()

# Categories
labels = [
    "Complaint",
    "Refund/Return",
    "Sales Inquiry",
    "Delivery Question",
    "Account/Technical Issue",
    "General Query",
    "Spam"
]

# AI Classification
def classify_category(text):
    result = classifier(text, labels)
    return result['labels'][0]

# Sentiment
def get_sentiment(text):
    result = sentiment_model(text)[0]
    if result['label'] == "POSITIVE":
        return "Positive"
    elif result['label'] == "NEGATIVE":
        return "Negative"
    else:
        return "Neutral"

# Reply Generator
def generate_reply(category, sentiment):

    if category == "Complaint":
        return "We sincerely apologize for the inconvenience. Please share more details so we can resolve this issue promptly."

    elif category == "Refund/Return":
        return "We understand your concern. Please initiate a return request and our team will assist you shortly."

    elif category == "Delivery Question":
        return "Thank you for reaching out. Kindly provide your order ID so we can check your delivery status."

    elif category == "Account/Technical Issue":
        return "We recommend trying to reset your credentials. If the issue persists, our support team is here to help."

    elif category == "Sales Inquiry":
        return "Thank you for your interest. Please let us know what product details or pricing information you need."

    elif category == "Spam":
        return "Your message appears to be unrelated. If you need assistance, feel free to contact support."

    else:
        return "Thank you for contacting us. We’re happy to assist you with your query."

# UI
user_input = st.text_area("Enter Customer Message")

if st.button("Process"):
    if user_input.strip():

        category = classify_category(user_input)
        sentiment = get_sentiment(user_input)
        reply = generate_reply(category, sentiment)

        st.subheader("Results")
        st.write(f"**Category:** {category}")
        st.write(f"**Sentiment:** {sentiment}")
        st.write(f"**Reply:** {reply}")
```
