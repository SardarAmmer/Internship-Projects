```python
# @title Task1 Umemployment Analysis, Exploring Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
df.head()
```

```python
# @title Renaming columns for easier use and checking any missing values
df.columns = [
    'State',
    'Date',
    'Frequency',
    'Unemployment_Rate',
    'Employed',
    'Labour_Participation_Rate',
    'Region',
    'Longitude',
    'Latitude'
]
# removong extra space
df['Date'] = df['Date'].astype(str).str.strip()
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df[['Date']].head()

# Check missing values
df.isnull().sum()
df.info()
```

```python
# @title Unemployment trend in Indian States Over Time
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6))
plt.plot(df['Date'], df['Unemployment_Rate'], alpha=0.4)
plt.title("Unemployment Rate Trend Over Time (All States)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

```python
# @title Unemployment trend in Indian States Over Time Region Wise
import matplotlib.pyplot as plt

region_avg = df.groupby('Region')['Unemployment_Rate'].mean()

plt.figure(figsize=(12,7))
plt.pie(region_avg, labels=region_avg.index, autopct='%1.1f%%', startangle=90)
plt.title("Region-wise Share of Average Unemployment")
plt.show()
```
